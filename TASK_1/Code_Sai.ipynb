# DEFINING DEPENDENT AND INDEPENDENT VARIABLES

```python
#Here x is  dependent (Excluding outcome column) variable and y as the independent variable (Outcome column)
import numpy as np
x=data.drop(columns='Outcome')
y=data['Outcome']
print(np.shape(x),np.shape(y))
#The outcomes decides whether the patient is diabetic or not. (1==>Diabetic 0==>Not)
y.value_counts()
```

# OUTLIER ANALYSIS

```python
#To find the presence of outliers and if present these outliers must be removed
#Here x is the dependent variable
# Plotting the data
style.use('classic')
fig, ax = plt.subplots()
ax.boxplot(x)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Box Plot')
plt.show()
```

#### BOX PLOT IS ONLY OF THE WAY TO DETECT TO PRESENCE OF OUTLIER

```python
# We need to remove the outlier,as they cause problem in statistical analysis
#Here we will be using IQR(Inter Quantile Range)
Q1 = x.quantile(0.25)
Q3 = x.quantile(0.75)
IQR = Q3 - Q1

# Remove the outliers which lies outside the bounds defined
x = x[~((x < (Q1 - 1.5 * IQR)) | (x > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
#The same data after the outliers are removed
style.use('')
fig, ax = plt.subplots()
ax.boxplot(x)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Box Plot')
plt.show()
#These ouliers needs to be removed,since the presence of them causes significant on the mean and standard deviation
```

# TRAIN AND TEST DATA SEPERATION

```python
#Train data and Test data split
X=data.drop(columns='Outcome')   
Y=data['Outcome']
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3)
```

```python
#Obviously this analysis falls under the classification category
#We need to find the best approach among the available classifications
#First loading all the required models
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
#We will be choosing which models suits it best with the help of accuracy score
#First storing all the available models in the dictionary name models
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": SVC()
}
```

```python
#We need to compute the accuracy score and based on it,we will decide which models suits this dataset more
accuracies={}                        #To store the model with accuracy values
for name, model in models.items():  
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies[name] = accuracy
    print(f"{name}: {accuracy}")
best_model = max(accuracies, key=accuracies.get)
print(f"The best model is {best_model}")
```

```python
#From the above approach we can conclude that Random forest is the best model 
```

```python
#Using grid_cv to perform hyper parameter tuning
from sklearn.model_selection import GridSearchCV

grid_parameters = {'n_estimators': [100, 200, 300, 400], 'max_depth': [5, 10, 15, 20]}
classifier = GridSearchCV(estimator=RandomForestClassifier(), param_grid=grid_parameters, cv=5)
classifier.fit(x_train, y_train)
```

```python
best_params = classifier.best_params_
print(best_params)
```

```python
best_score = classifier.best_score_
print(best_score)
```

```python
print(classifier.score(x_test,y_test))
y_predict=classifier.predict(x_test)
y_predict
```

```python
#Confusion matrix can be used to evalate the model
from sklearn.metrics import confusion_matrix
```

```python
conf_matrix=confusion_matrix(y_test,y_predict)
print(conf_matrix)
```

### (0,0)=>TRUE POSITIVE  
### (0,1)=>FALSE POSITIVE 
### (1,0)=>FALSE NEGATIVE
### (1,1)=>TRUE NEGATIVE

```python
#To visualize this we can use seaborn library
sbn.heatmap(conf_matrix,annot=True,fmt='g')
plt.xlabel('Predicted_label')
plt.ylabel('True_label')
plt.title('Confusion_matrix')
```
