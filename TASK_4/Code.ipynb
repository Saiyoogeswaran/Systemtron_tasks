# OUTLIERS ANALYSIS

```python
#Let's visualize the data with boxplot to find whether outliers are present or not
from matplotlib import style 
plt.figure(figsize=(20, 10))
from matplotlib import style 
style.use('classic')
plt.boxplot(X)
# Set the x-axis labels
plt.xticks(range(1, len(X.columns) + 1), X.columns)
plt.xticks(rotation=75)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Box Plot')
plt.show()
```

```python
data.columns
```

```python
# Select the columns to remove outliers from
cols = ['area_mean','area_se', 'area_worst']

# Calculate the first and third quartiles
Q1 = X[cols].quantile(0.25)
Q3 = X[cols].quantile(0.75)

# Calculate the interquartile range
IQR = Q3 - Q1

# Remove the outliers
X = X[~((X[cols] < (Q1 - 3* IQR)) | (X[cols] > (Q3 + 3* IQR))).any(axis=1)]
```

```python
#Let's visualize the data with boxplot to find whether outliers are present or not
from matplotlib import style 
plt.figure(figsize=(20, 10))
from matplotlib import style 
style.use('classic')
plt.boxplot(X)
# Set the x-axis labels
plt.xticks(range(1, len(X.columns) + 1), X.columns)
plt.xticks(rotation=75)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Box Plot')
plt.show()
```

# TEST AND TRAIN DATA SEPERATION

```python
from sklearn.model_selection import train_test_split
X=data.drop('diagnosis',axis='columns')
y=data['diagnosis']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
#We need to find the best approach among the available classifications
#First loading all the required models
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
#We will be choosing which models suits it best with the help of accuracy score
#First storing all the available models in the dictionary name models
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": SVC()
}
```

```python
#We need to compute the accuracy score and based on it,we will decide which models suits this dataset more
accuracies={}                        #To store the model with accuracy values
for name, model in models.items():  
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies[name] = accuracy
    print(f"{name}: {accuracy}")
best_model = max(accuracies, key=accuracies.get)
print(f"The best model is {best_model}")
```

# MODEL PREDICTION

### LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
# Create an instance of the LogisticRegression class
lr = LogisticRegression()

# Define the hyperparameter grid
param_grid = {'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
              'penalty': ['l1', 'l2', 'elasticnet'],
              'C': [ 0.1, 1, 10]}
```

```python
# Create an instance of the GridSearchCV class
grid = GridSearchCV(lr, param_grid, cv=5)

# Fit the model to the data
grid.fit(X_train, y_train)

# Print the best hyperparameters
print("Best hyperparameters:", grid.best_params_)

# Print the best score
print("Best score:", grid.best_score_)

#Store the best estimator as best_lr 
best_lr = grid.best_estimator_
```

### Best hyperparameters: {'C': 10, 'penalty': 'l1', 'solver': 'liblinear'}
### Best score: 0.9798734177215189

```python
print(best_lr.score(X_test,y_test))
y_predict=best_lr.predict(X_test)
y_predict
```
